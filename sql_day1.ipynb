```python
import sqlite3
import pandas as pd

# Create a database connection
conn = sqlite3.connect('practice.db')
print("Database connected successfully!")
```

```python
df = pd.read_csv('C:/pd/superstore.csv', encoding='latin1')
df.to_sql('superstore', conn, if_exists='replace', index=False)

print("Data loaded into database successfully!")
print(f"Rows loaded: {len(df)}")
```

```python
query = """
SELECT *
FROM superstore
LIMIT 5
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT Region, Sales, Profit
FROM Superstore
LIMIT 10
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT Region, Sales, Profit
FROM Superstore
WHERE Region = 'West'
LIMIT 10
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT COUNT(*)  AS total_rows FROM superstore
"""
result = pd.read_sql_query(query, conn)
result
```

```python
#region wise sales
query = """
SELECT Region, SUM(Sales) AS total_sales
FROM superstore
GROUP BY Region
"""
result = pd.read_sql_query(query, conn)
result
```

```python
#AVG profit
query = """
SELECT Category, AVG(Profit) AS AVG_PROFIT
FROM superstore
GROUP BY Category
"""

result = pd.read_sql_query(query, conn)
result
```

```python
#total sales

query = """
SELECT "Customer Name", SUM(Sales) AS total_sales
FROM superstore
GROUP BY "Customer Name" 
ORDER BY total_sales DESC
LIMIT 5;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
#Total profit by category and region is west
query = """
SELECT Category, SUM(Profit) AS TOTAL_PROFIT
FROM superstore
WHERE Region = 'West'
GROUP BY Category
ORDER BY TOTAL_PROFIT DESC
"""

result = pd.read_sql_query(query, conn)
result
```
