```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('practice.db')

# Create customers table
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Shruti', 'Raj', 'Priya', 'Arjun', 'Meera'],
    'city': ['Surat', 'Mumbai', 'Delhi', 'Pune', 'Ahmedabad']
})

# Create orders table
orders = pd.DataFrame({
    'order_id': [101, 102, 103, 104],
    'customer_id': [1, 2, 1, 6],
    'amount': [500, 1200, 800, 300]
})

# Load into database
customers.to_sql('customers', conn, if_exists='replace',index='False')
orders.to_sql('orders', conn, if_exists='replace', index='False')

print("Tables created successfully!")
print("\nCustomers table:")
print(customers)
print("\nOrders table:")
print(orders)
```

```python
query = """
SELECT customers.name, customers.city, orders.order_id, orders.amount
FROM "customers"
INNER JOIN "orders" ON "customers".customer_id = "orders".customer_id
"""

result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT customers.name, customers.city, orders.order_id, orders.amount
FROM "customers"
LEFT JOIN "orders" ON "customers".customer_id = "orders".customer_id
"""

result = pd.read_sql_query(query, conn)
result
```

```python
# Split superstore into two tables
df = pd.read_csv('C:/pd/superstore.csv', encoding='latin1')

# Table 1 — orders info
orders_df = df[['Order ID', 'Customer Name', 'Region', 'Sales', 'Profit']]

# Table 2 — product info  
products_df = df[['Order ID', 'Category', 'Sub-Category', 'Product Name']]

# Load into database
orders_df.to_sql('orders_info', conn, if_exists='replace', index=False)
products_df.to_sql('products_info', conn, if_exists='replace', index=False)

print("Tables created!")
print("Orders table shape:", orders_df.shape)
print("Products table shape:", products_df.shape)
```

```python
query = """
SELECT products_info.Category, SUM(orders_info.Sales) AS total_sales
FROM orders_info
INNER JOIN products_info ON orders_info."Order ID" = products_info."Order ID"
GROUP BY Category
ORDER BY total_sales DESC
"""
#here i have found duplicate join problem
result = pd.read_sql_query(query, conn)
result
```
